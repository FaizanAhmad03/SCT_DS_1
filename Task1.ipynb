```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np


csv_path = "synthetic_cancer_patients_6types.csv"
df = pd.read_csv(csv_path)


print(df.head())

cancer_counts = df["cancer_type"].value_counts()

print("Cancer type counts:")
print(cancer_counts)


counts_df = cancer_counts.reset_index()
counts_df.columns = ["cancer_type", "count"]


counts_df = counts_df.sample(frac=1, random_state=None).reset_index(drop=True)

color_palette = ["#FF6B6B", "#4ECDC4", "#FFD93D", "#1A535C", "#FF9F1C", "#9B5DE5"]
colors = color_palette[:len(counts_df)]

plt.style.use("seaborn-v0_8-darkgrid")

plt.figure(figsize=(9, 5))
bars = plt.bar(
    counts_df["cancer_type"],
    counts_df["count"],
    color=colors,
    edgecolor="black",
    linewidth=1.5
)

for bar in bars:
    height = bar.get_height()
    plt.text(
        bar.get_x() + bar.get_width() / 2,
        height + 5,
        str(height),
        ha="center",
        va="bottom",
        fontsize=9,
        fontweight="bold"
    )


plt.xlabel("Cancer Type", fontsize=11, fontweight="bold")
plt.ylabel("Number of Patients", fontsize=11, fontweight="bold")
plt.title("Distribution of Cancer Types Post Covid-19", fontsize=13, fontweight="bold")


plt.xticks(rotation=20, ha="right", fontsize=10)

plt.tight_layout()
plt.show
```

```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from google.colab import files
import io


print("Please upload your 'synthetic_traffic_accidents.csv' file:")
uploaded = files.upload()


filename = list(uploaded.keys())[0]
accidents_df = pd.read_csv(io.BytesIO(uploaded[filename]))

print(f"âœ… Dataset '{filename}' loaded successfully!")
print(f"Dataset shape: {accidents_df.shape}")
print("\nFirst 5 rows:")
print(accidents_df.head())
print("\nDataset columns:", accidents_df.columns.tolist())


sns.set(style="whitegrid")


plt.figure(figsize=(10,6))
sns.countplot(data=accidents_df, x='Road_Condition', order=accidents_df['Road_Condition'].value_counts().index)
plt.title('Accidents by Road Condition', fontsize=14, fontweight='bold')
plt.xlabel('Road Condition', fontsize=12)
plt.ylabel('Number of Accidents', fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()


plt.figure(figsize=(10,6))
sns.countplot(data=accidents_df, x='Weather', order=accidents_df['Weather'].value_counts().index)
plt.title('Accidents by Weather Condition', fontsize=14, fontweight='bold')
plt.xlabel('Weather', fontsize=12)
plt.ylabel('Number of Accidents', fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

plt.figure(figsize=(10,6))
sns.countplot(data=accidents_df, x='Time_of_Day', order=accidents_df['Time_of_Day'].value_counts().index)
plt.title('Accidents by Time of Day', fontsize=14, fontweight='bold')
plt.xlabel('Time of Day', fontsize=12)
plt.ylabel('Number of Accidents', fontsize=12)
plt.tight_layout()
plt.show()


plt.figure(figsize=(12,6))
location_counts = accidents_df['Location'].value_counts()
sns.barplot(x=location_counts.index, y=location_counts.values, palette='viridis')
plt.title('Accident Hotspots by Location', fontsize=14, fontweight='bold')
plt.xlabel('Location', fontsize=12)
plt.ylabel('Number of Accidents', fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()


plt.figure(figsize=(12,6))
sns.countplot(data=accidents_df, x='Contributing_Factor', order=accidents_df['Contributing_Factor'].value_counts().index, palette='Set2')
plt.title('Contributing Factors to Accidents', fontsize=14, fontweight='bold')
plt.xlabel('Contributing Factor', fontsize=12)
plt.ylabel('Number of Accidents', fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()


plt.figure(figsize=(10,6))
sns.countplot(data=accidents_df, x='Severity', order=accidents_df['Severity'].value_counts().index, palette='Reds_r')
plt.title('Accident Severity Distribution', fontsize=14, fontweight='bold')
plt.xlabel('Severity', fontsize=12)
plt.ylabel('Number of Accidents', fontsize=12)
plt.tight_layout()
plt.show()

print("ðŸŽ‰ All visualizations completed successfully!")
```
